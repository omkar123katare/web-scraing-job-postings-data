**Importing the necessary libraries.**

```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

**Extract Function to extract soup(raw data).**

```python

def extract(page):
  ''' 
  This function takes in the variable part of web page url and goes through the webpage of corresponding url to get raw data.
  '''
  headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/97.0.4692.99 Safari/537.36'}
  url=f"https://in.indeed.com/jobs?q=data%20analyst&l=India&start={page}"
  r=requests.get(url,headers)
  #return r.status_code
  soup=BeautifulSoup(r.content,'html.parser')
  return soup
```

**Transform function to transform the data.**

```python

def transform(onm):
  '''
  This function takes in the raw data from extract function and by making use of .find_all() we find the required information using the tags.
  '''
  role=[]
  for elem in onm.find_all('span'):
    role.append(elem.attrs)

  for a in role:
    if 'title' in list(a.keys()):
      for elem in list(a.values()):
        job_role.append(elem)

  for elem in onm.find_all("span",class_="companyName"):
    company_name.append(elem.text)

  for elem in onm.find_all("div",class_="companyLocation"):
    job_location.append(elem.text)

  for elem in onm.find_all("span",class_="date"):
    try:
      post_date.append(elem.text)
    except:
      post_date.append('N/A')

  rating=[]
  for elem in onm.find_all('span'):
    try:
      rating.append(elem.attrs)
    except:
      rating.append('N/A')

  for elem in rating:
    if "aria-label" in list(elem.keys()):
      company_rating.append(str(list(elem.values())[1]))

  return(job_role,job_location,company_name,company_rating,post_date)
```

**Now getting the raw data from the web pages using for loop to scrape through 600 pages**

```python
role=[]
company_name=[]
job_location=[]
post_date=[]
rating=[]
job_role=[]
company_rating=[]


for i in range(0,6000,10):
  extraction_phase=extract(i)
  a,b,c,d,e=transform(extraction_phase)


jobs_dataframe=pd.DataFrame([c,a,b,d,e])
```
